# 作图的代码

```python
r1 = [0.667, 0.733, 0.816, 0.765, 0.706, 0.5, 0.632, 0.571]
r2 = [0.339, 0.312, 0.333, 0.222, 0.311, 0.385, 0.326, 0.309]

data = {
    'Wrist error': [r1[0], r2[0]],
    'Arm error': [r1[1], r2[1]],
    'Racket angle error': [r1[2], r2[2]],
    'Hitting position error': [r1[3], r2[3]],
    'Timing error':[r1[4], r2[4]],
    'Footwork lag error': [r1[5], r2[5]],
    'Slow body transition error': [r1[6], r2[6]],
    'High center of gravity error': [r1[7], r2[7]]
}

# 提取数据
labels = list(data.keys())
accuracy = [item[0] for item in data.values()]
false_positive_rate = [item[1] for item in data.values()]

# 设置柱状图的位置和宽度
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, accuracy, width, label='Accuracy')
rects2 = ax.bar(x + width/2, false_positive_rate, width, label='False Positive Rate')

# 添加文本标签、标题和自定义X轴刻度标签
ax.set_ylabel('Scores')
ax.set_ylim([0, 1])
ax.set_title('Accuracy and False Positive Rate by Error Type')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.legend()

# 为每个柱子添加数值标签
def autolabel(rects):
    """在每个柱子上方添加一个文本标签，显示其高度。"""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height,2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.savefig('fig5.jpg', dpi=300)
plt.show()
```
